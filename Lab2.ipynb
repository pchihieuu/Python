```python
'''
Viết chương trình nhập số nguyên n, in ra kết quả như hình sau:  Nhập số n: 6
6 5 4 3 2 1
5 4 3 2 1
4 3 2 1
3 2 1
2 1
1
'''
# @author: Hieu Phan Chi (@aka: pchihieu)
# Định nghĩa hàm in theo thiết kế trên
def print_pattern(n):
    for i in range(n, 0, -1):
        for j in range(i, 0, -1):
            if j == 1:
                print(j)
            else:
                print(j, end=' ')
        if i > 1:
            print()
# Kiểm tra xem câu lệnh có chạy trong chương trình chính hay không?
if __name__=="__main__":
  # Nhập số n
  n = int(input("Nhập số n: "))
  # In kết quả
  print_pattern(n)
```

```python
'''
Viết chương trình nhập số nguyên n, in ra các số nguyên tố từ 1 đến <=n.
'''
# @author: Hieu Phan Chi (@aka: pchihieu)
# Định nghĩa hàm sử dụng thuật toán sàng số nguyên tố
# bigO: O(nlog)
def sieve_of_eratosthenes(n):
  # Khởi tạo mảng theo dỗi số nguyên tố hay không? đầu tiên gán nó là SNT
    is_prime = [True] * (n + 1)
    p = 2
    '''
    Kiểm tra từng số nguyên tố lần lượt
    Kiểm tra xem p có là SNT không? nếu True thì loại bỏ bội số của p
    '''
    while p ** 2 <= n:
        if is_prime[p]:
          # Lập trên tập p, step: p. Các số không phải SNT được đánh dầu là False trong mảng is_prime
            for i in range(p ** 2, n + 1, p):
                is_prime[i] = False
        # Tăng p lên 1 để kiểm tra SNT tiếp theo
        p += 1
    # True: SNT , False: không phải SNT
    # Sử dụng mảng is_prime để tạo primes chứ các SNT từ 2 đến n
    primes = [p for p in range(2, n + 1) if is_prime[p]]
    # Trả về giá trị primes
    return primes
# Kiểm tra xem câu lệnh có chạy trong chương trình chính hay không?
if __name__=="__main__":
  # Nhập giá trị n
    n = int(input("Nhập số nguyên n: "))
    # Gán giá trị prime_numers bằng hàm sieve_of_eratosthenes(n)
    prime_numbers = sieve_of_eratosthenes(n)
    # In kết quả ra màn hình
    print("Các số nguyên tố từ 1 đến", n, "là:")
    print(prime_numbers)
```

```python
'''
Viết chương trình nhập vào 1 list, thực hiện:
 + Xuất List cho xem
 + Thêm một phần tử và list,
 + Xóa 1 phần tử khỏi List,
 + Đếm phần tử x có trong List bao nhiêu lần
 + Đảo ngược List,
 + Sắp xếp List tăng, giảm dần
'''
# @author: Hieu Phan Chi (@aka: pchihieu)
# using SOLID
# Định nghĩa lớp listManager
class listManager:
    # Định nghĩa hàm khởi tạo danh sách
    def __init__(self):
        self.my_list = []
    # Định nghĩa hàm in danh sách
    def display_list(self):
        print("Danh sách hiện tại:", self.my_list)
    # Định nghĩa hàm thêm phần tử vào danh sách
    def add_element(self, element):
        self.my_list.append(element)
    # Định nghĩa hàm xóa phần tử trong danh sách
    def remove_element(self, element):
        if element in self.my_list:
            self.my_list.remove(element)
        else:
            print(f"{element} không tồn tại trong danh sách.")
    # Định nghĩa hàm đếm phần tử xuất hiện bao nhiêu lần trong danh sách
    def count_element(self, element):
        count = self.my_list.count(element)
        print(f"{element} xuất hiện {count} lần trong danh sách.")
    # Định nghĩa hàm đảo ngược danh sách
    def reverse_list(self):
        self.my_list = self.my_list[::-1]
    # Định nghĩa hàm sắp xếp phần tử tăng dần
    def sort_list(self, reverse=False):
        self.my_list.sort(reverse=reverse)
# Định nghĩa hàm main để tạo Menu vào thực thi chương trình
def main():
  # Khởi tạo đối tượng bằng class listManager
    manager = listManager()
    # In menu ra màn hình
    while True:
        print("Chọn tùy chọn:")
        print("1. Xuất danh sách")
        print("2. Thêm phần tử")
        print("3. Xóa phần tử")
        print("4. Đếm phần tử")
        print("5. Đảo ngược danh sách")
        print("6. Sắp xếp danh sách (tăng dần)")
        print("7. Sắp xếp danh sách (giảm dần)")
        print("8. Thoát")
        # Khởi tạo biến để nhập lựa chọn
        choice = input("Nhập lựa chọn: ")

        if choice == '1':
            manager.display_list()
        elif choice == '2':
            element = input("Nhập phần tử muốn thêm: ")
            manager.add_element(element)
        elif choice == '3':
            element = input("Nhập phần tử muốn xóa: ")
            manager.remove_element(element)
        elif choice == '4':
            element = input("Nhập phần tử muốn đếm: ")
            manager.count_element(element)
        elif choice == '5':
            manager.reverse_list()
        elif choice == '6':
            manager.sort_list()
        elif choice == '7':
            manager.sort_list(reverse=True)
        elif choice == '8':
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
# Kiểm tra xem câu lệnh có chạy trong chương trình chính hay không?
if __name__ == "__main__":
    main()
```

```python
'''
Viết chương trình nhập vào 1 list, chuyển list thành tuple và thực hiện:
 + Xuất Tuple cho xem, Thêm 1 phần tử vào Tuple,
 + Xóa 1 phần tử khỏi Tuple: Không thể xóa 1 phần tử khỏi Tuple, nhưng có thể xóa hoàn toàn Tuple,
 + Đếm phần tử x có trong Tuple bao nhiêu lần,
 + Cho biết số phần tử của Tuple
'''
# @author: Hieu Phan Chi (@aka: pchihieu)
# using SOLID
# Định nghĩa hàm tạo danh sách tuple, cách nhau bằng dầu ','
def create_tuple():
    user_input = input("Nhập các phần tử vào danh sách: ")
    user_list = [item.strip() for item in user_input.split(",")]
    user_tuple = tuple(user_list)
    return user_tuple
# Định nghĩa hàm in danh sách của Tuple
def display_tuple(user_tuple):
    print("Danh sách của tuple: ", user_tuple)
# Định nghĩa hàm thêm phần tử vào Tuple
def add_element(user_tuple, element):
    new_tuple = user_tuple + (element,)
    return new_tuple
# Định nghĩa hàm xóa Tuple
def delete_tuple(user_tuple):
    del user_tuple
    return None
# Định nghĩa hàm đếm phần tử xuất hiện trong Tuple
def count_element(user_tuple, element):
    count = user_tuple.count(element)
    print(f"{element} xuất hiện {count} lần trong danh sách tuple")
# Định nghĩa hàm main để chạy toàn bộ chương trình
def main():
  # Khởi tạo đối tượng tuple
    user_tuple = create_tuple()
    # In Menu
    while True:
        print("Menu tùy chọn:")
        print("1. Xuất Tuple")
        print("2. Thêm phần tử vào Tuple")
        print("3. Xóa phần tử khỏi Tuple")
        print("4. Đếm phần tử trong Tuple")
        print("5. Số phần tử trong Tuple")
        print("6. Thoát chương trình")
        # Tạo biến chọn
        choice = input("Nhập một trong số lựa chọn trên: ")
        if choice == '1':
            display_tuple(user_tuple)
        elif choice == '2':
            element = input("Nhập phần tử muốn thêm: ")
            user_tuple = add_element(user_tuple, element)
            display_tuple(user_tuple)
        elif choice == '3':
            user_tuple = delete_tuple(user_tuple)
            print("Tuple đã bị xóa.")
        elif choice == '4':
            element = input("Nhập phần tử muốn đếm: ")
            count_element(user_tuple, element)
        elif choice == '5':
            count = len(user_tuple)
            print(f"Số phần tử trong Tuple: {count}")
        elif choice == '6':
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng chọn lại")
# Kiểm tra xem câu lệnh có chạy trong chương trình chính hay không?
if __name__ =="__main__":
  main()
```

```python
'''
Viết chương trình nhập vào 2 set: set A, set B, thực hiện tính:
Tập A – B
Tập B – A
Tập A ∩ B
Tập A ∪ B
'''
# @author: Hieu Phan Chi (@aka: pchihieu)
# using SOLID
# Định nghĩa lớp setManager
class setManager:
  # Định nghĩa hàm set
    def __init__(self, set_a, set_b):
        self.set_a = set_a
        self.set_b = set_b
    # Định nghĩa hàm tính A - B
    def difference_a_b(self):
        return self.set_a - self.set_b
    # Định nghĩa hàm tính B - A
    def difference_b_a(self):
        return self.set_b - self.set_a
    # Định  nghĩa hàm tính A ∩ B
    def intersection(self):
        return self.set_a & self.set_b
    # Định nghĩa hàm tính A ∪ B
    def union(self):
        return self.set_a | self.set_b
# Định nghĩa hàm nhập phần từ vào tập hợp
def create_set_from_input():
    user_input = input("Nhập các phần tử của tập hợp, cách nhau bằng dấu phẩy: ")
    user_list = [item.strip() for item in user_input.split(",")]
    user_set = set(user_list)
    return user_set
# Định nghĩa hàm hiển thị
def display_set(title, s):
    print(f"{title}:", s)
# Định nghĩa hàm main để thực thi toàn bộ chương trình
def main():
    # Nhập giá trị hai tập hợp A và B
    print("Chương trình thực hiện phép toán trên hai tập hợp A và B.")
    set_a = create_set_from_input()
    set_b = create_set_from_input()
    # Khởi tạo đối tượng thuộc lớp setManger
    manager = setManager(set_a, set_b)
    '''
    # Not using Menu
    display_set("Tập A", set_a)
    display_set("Tập B", set_b)

    result_a_minus_b = manager.difference_a_b()
    result_b_minus_a = manager.difference_b_a()
    result_intersection = manager.intersection()
    result_union = manager.union()

    display_set("Tập A - B", result_a_minus_b)
    display_set("Tập B - A", result_b_minus_a)
    display_set("Tập A ∩ B", result_intersection)
    display_set("Tập A ∪ B", result_union)
    '''
    # In menu
    while True:
        print("\nChọn tùy chọn:")
        print("1. Hiển thị tập hợp A")
        print("2. Hiển thị tập hợp B")
        print("3. Tập A - B")
        print("4. Tập B - A")
        print("5. Tập A ∩ B")
        print("6. Tập A ∪ B")
        print("7. Thoát")
        # Tạo biến chọn
        choice = input("Nhập lựa chọn: ")

        if choice == '1':
            display_set("Tập A", set_a)
        elif choice == '2':
            display_set("Tập B", set_b)
        elif choice == '3':
            result = manager.difference_a_b()
            display_set("Tập A - B", result)
        elif choice == '4':
            result = manager.difference_b_a()
            display_set("Tập B - A", result)
        elif choice == '5':
            result = manager.intersection()
            display_set("Tập A ∩ B", result)
        elif choice == '6':
            result = manager.union()
            display_set("Tập A ∪ B", result)
        elif choice == '7':
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
# Kiểm tra xem câu lệnh có chạy trong chương trình chính hay không?
if __name__ == "__main__":
    main()
```

```python
'''
Viết chương nhập 2 số nguyên a, b, xác định ước số chung lớn nhất của a, b
'''
#@author: Hieu Phan Chi(@aka: pchihieu)
# Định nghĩa hàm tìm USCLN bằng Euclid
def find_gcd(a, b):
    while b:
        a, b = b, a % b
    return a
# Định nghĩa hàm main để thực thi toàn bộ chương trình
def main():
    # Sử dụng xử lý ngoại lệ exception handling
    try:
        # Nhập giá trị hai số a và b
        a = int(input("Nhập số nguyên a: "))
        b = int(input("Nhập số nguyên b:"))

        if a < 0 or b < 0:
            print("Vui lòng nhập hai số nguyên không âm.")
        else:
            gcd = find_gcd(a, b)
            print(f"Ước số chung lớn nhất của {a} và {b} là {gcd}")
    except ValueError:
        print("Vui lòng nhập hai số nguyên hợp lệ.")

if __name__ == "__main__":
    main()
```

```python
'''
Viết chương nhập số nguyên n, tính n!
'''
# @author: Hieu Phan Chi(@aka:pchihieu)
# Định nghĩa hàm tinh toán giai thừa của số n
def calculate_factorial(n):
    if n < 0:
        return "Không thể tính giai thừa cho số âm."
    elif n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result
# Định nghĩa hàm main để thực thi toàn bộ chương trình
def main():
    # Sử dụng xử lý ngoại lệ exception handling
    try:
        n = int(input("Nhập một số nguyên n: "))
        factorial = calculate_factorial(n)
        print(f"{n}! = {factorial}")
    except ValueError:
        print("Vui lòng nhập một số nguyên hợp lệ.")

if __name__ == "__main__":
    main()
```

```python
'''
Viết chương trình tính giá trị số fibonacci thứ n
'''
#@author: Hieu Phan Chi(@aka: pchihieu)
# Định nghĩa hàm tìm số fibonacci bằng cách sử dụng vòng lặp for( thay vì đệ quy)
def fibonacci_iterative(n):
    if n <= 0:
        return "Không có số Fibonacci nào cho n <= 0."
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        a, b = 0, 1
        for _ in range(2, n):
            a, b = b, a + b
        return b
# Định nghĩa hàm main để thực thi toàn bộ chương trình
def main():
    # Sử dụng xử lý ngoài lệ exception handling
    try:
        n = int(input("Nhập một số nguyên dương n: "))
        if n < 1:
            print("Vui lòng nhập một số nguyên dương.")
        else:
            result = fibonacci_iterative(n)
            print(f"Số Fibonacci thứ {n} là {result}")
    except ValueError:
        print("Vui lòng nhập một số nguyên hợp lệ.")

if __name__ == "__main__":
    main()
```

```python
'''
Viết chương trình nhập vào số nguyên n, tính S = 1 + ½ + 1/3 + ..+ 1/n và in kết quả
'''
# @author: Hieu Phan Chi(@aka: pchihieu)
# Định nghĩa hàm tính tổng
def sum_number(n):
  # Tạo biến S đầu tiên bằng 0
  s = 0.0
  # Sử dụng vòng lập để duyệt trên tập n + 1
  for i in range(1, n + 1):
    # giá trị biểu thức trên được tính theo công thức sau:
    s += 1/i
  return s
# Kiểm tra câu lệnh có chạy trong chương trình main hay không?
if __name__ == "__main__":
  # Nhập giá trị n
  n = int(input("Nhập số n: "))
  # In kết quả ra màn hình
  print("Tổng dãy số bằng: ", sum_number(n))
```

```python
'''
Viết chương trình nhập số tiến hiện có x, nhập số tiền mơ ước y (y>x), thực hiện gởi
ngân hàng với lãi suất 3% / 1 tháng, hỏi phải mất bao nhiêu tháng gởi để có được số tiền y
'''
# @author: Hieu Phan Chi(@aka: pchihieu)
# Định nghĩa hàm tính tháng để đạt được số tiền mơ ước
def money_month(x, y):
  # Tạo biến month = 0
  month = 0
  # Sử dụng vòng lập while để kiểm tra số tiền hiện có và mơ ước hợp lệ
  while(x < y):
    # Lãi xuất hàng tháng 3%
    x += x * 0.03
    # Tính toán số tháng thỏa mãn
    month += 1
  # Trả về giá trị tháng
  return month
# Kiểm tra câu lệnh có chạy trong chương trình chính hay không?
if __name__ == "__main__":
  # Xử lý ngoại lệ
  try:
    # Nhập các giá trị tiền hiện có và tiền mơ ước
    x = float(input("Nhập số tiền hiện có: "))
    y = float(input("Nhập số tiền mơ ước: "))
    # Kiểm tra điều kiện ngoại lệ và thông báo lỗi
    if x >= y:
      raise ValueError("Số tiền hiện có phải bé hơn số tiền mơ ước!")
    # In kết quả ra màn hình
    print(f"Cần {money_month(x, y)} tháng gởi để có được số tiền {y}")
  # Phát hiện ngoại lệ cho điều kiện x >= y
  except ValueError as e:
    print(f"Lỗi: {e}")
  # Xử lý ngoại lệ không xác định
  except Exception as a:
    print(f"Lỗi không xác định được: {a}")
```
